# CODSOFT TASK-3 IRIS FLOWER CLASSIFICATION

## Libraries and Dataset

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

iris = pd.read_csv("task3.csv")
```

# The shape of dataset

```python
iris.shape
```

## Explore the information and dataset

```python
iris.info()
```

## Independent Variables

```python
X = iris.iloc[ : , 0:4]
X
```

## Extracting Dependent Variable

```python
Y = iris.iloc[ : , 4: ]
Y.species.unique()
```

## Implementing CART Algorithm for DTL

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X , 
                                                   Y, 
                                                   test_size = 0.25, 
                                                   random_state = 5)
X_train
```

## construct the DT model

```python
from sklearn.tree import DecisionTreeClassifier 

clf = DecisionTreeClassifier(random_state = 1234, criterion = 'entropy')

clf.fit(X_train , Y_train)
```

## Visualize the decision tree as text

```python
from sklearn import tree

text_representation = tree.export_text(clf)

print(text_representation)
```

# Decision Tree Pictorially


```python
fig = plt.figure(figsize = (25 , 20) , dpi = 200.0)

_ = tree.plot_tree(clf, 
                   feature_names = ['sepal.length' ,'sepal.width', 'petal.length', 'petal.width'],
                   class_names = ['setosa', 'versicolor', 'virginica'],
                   filled = True)
```

# Testing the accuracy of the DT model

```python
from sklearn.metrics import accuracy_score

pred_train = clf.predict(X_train)

accuracy_train = accuracy_score(Y_train, pred_train)

print('% of Accuracy on training data: ', accuracy_train * 100 )



# Let us test the accuracy of the model on the test data (or new data or unseen data).

pred_test = clf.predict(X_test)

accuracy_test = accuracy_score(Y_test, pred_test)

print('% of Accuracy on test data: ', accuracy_test * 100 )
```
